```python
from dendrogram_tools import dend_json_2_nodes_n_edges
tree = dend_json_2_nodes_n_edges('../src/dendrograms/CCN202002013.json')
tree['nodes'][0]
```

```python
records = []
for n in tree['nodes']:
    parents = [e[1] for e in tree['edges'] if e[0] == n['cell_set_accession']]
    n['parents'] = '|'.join(parents) # Hmmm - doesn't everything have only one parent?
    records.append(n)
```

```python
import pandas as pd
```

```python
tab = pd.DataFrame.from_records(records)
tab[0:5]
```
